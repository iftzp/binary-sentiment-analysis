```python
import pandas as pd
```

```python
data_path = "data/subjectivity_clues_hltemnlp05/subjclueslen1-HLTEMNLP05.tff"
```

```python
data = {}

with open(data_path, 'r') as f:
    while len(f.readline()) != 0:
        toks = f.readline().split()
        
        word = toks[2].split("=")[1]
        polarity = toks[-1].split("=")[1]
        
        if polarity in ("positive", "negative"):
            data[word] = 1 if polarity == "positive" else 0
```
